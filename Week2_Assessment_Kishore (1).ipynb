Q1. What's the difference between a Series and a DataFrame? Give one operation that works on both,
and one that works only on a DataFrame.

Answer: A Series is a 1-dimensional labeled array capable of holding data of any type, whereas a DataFrame is a 2-dimensional tabular data structure with rows and columns b.Operation on both: .head() or .isnull() works perfectly on both structures.



Q2. What does inplace=True do? Why is the pandas team discouraging its use in newer versions? Give
one example where you'd avoid it.

Answer: inplace=True modifies the underlying object directly rather than returning a new modified copy.

Discouragement Reason: The pandas team discourages it because it does not actually save memory as commonly believed; it often triggers internal layout copies anyway, creates confusing chained assignment patterns, and blocks functional method-chaining workflows.

Example to avoid: df.dropna(inplace=True) should be avoided; use df = df.dropna() instead.

Q3. Explain the difference between .fillna() and .dropna(). When would a data analyst pick one over the
other? Be specific - give a concrete situation for each.

Answer: .dropna() removes entire rows or columns containing missing (NaN) values, while .fillna() replaces missing values with a specified default or computed metric (like the mean, median, or mode).

When to pick .dropna(): In a medical dataset where critical target metrics (like blood pressure values) are missing entirely, inventing data via filling might distort analysis.

When to pick .fillna(): In transactional records where a missing discount_amount simply means no discount was applied; here, filling with 0 preserves structural transaction rows.

Q4. What does pd.to_numeric(col, errors='coerce') do that plain pd.to_numeric(col) does NOT? Which
one would you use on real-world messy data, and why?

Answer: Plain pd.to_numeric(col) throws a hard runtime exception if it encounters any non-numeric text string  , halting script execution. Setting errors='coerce' gracefully converts those invalid textual items into NaN values without breaking execution.
Choice: You should always choose errors='coerce' on real-world messy data because dirty data streams contain unexpected text snippets that require safe pipeline ingestion.

Q5. Explain .loc vs .iloc with one example each. Then describe a case where df.loc[0] and df.iloc[0]
return different rows.

Answer: .loc uses label-based indexing (matching index names/labels), while .iloc uses integer position-based indexing (matching structural row sequences from 0 to $N-1$).
Example .loc: df.loc[10] finds the row labeled 10.
Example .iloc: df.iloc[10] finds the 11th literal row in the DataFrame.
Divergence Case: If a DataFrame is sorted or shuffled (e.g., via df.sort_values()), the row at the absolute top of the table might have a label like 42. In this case, df.iloc[0] fetches row label 42 (the structural first), while df.loc[0] looks for the row where the original index label is explicitly 0 (which could now be at the bottom).

Q6. What is the difference between .map(), .apply(), and .replace() for a Series? When would you reach
for each?


Answer:

.map() is best suited for element-wise transformations on a Series using a dictionary mapping or a single-argument function.

.apply() runs a custom function across each individual element of a Series (or passes complex function objects).

.replace() substitutes specific target values with new values across a Series or DataFrame without needing to specify a map for the other unchanged elements.

Usage: Reach for .map() for standard categorization dictionaries; reach for .apply() for dynamic numeric modifications (like lambda scaling calculations); reach for .replace() to clean specific garbage strings.

Q7. What is the SettingWithCopyWarning? Give one example of code that triggers it and one way to fix
it.  

Answer: It is a warning triggered when pandas cannot guarantee whether a value-assignment operation modified the original data array or an isolated temporary data view.

Triggers it: df_filtered = df[df['rating'] > 4] followed by df_filtered['rating'] = 10 (chained assignment).

Fix: Use .copy() during creation: df_filtered = df[df['rating'] > 4].copy().

Q8. Explain pd.cut() vs pd.qcut(). If you have annual income data and want to label customers as Low /
Mid / High earners, which one fits better? Why?

Answer: pd.cut() segments data into equal-width bins based on geometric range values, whereas pd.qcut() splits data into equal-size quantiles (meaning each bin gets approximately the same total number of data rows).

Choice: Use pd.cut() for categorizing income earners. Wealth distributions are naturally highly skewed (a few billionaires can push thresholds out). pd.cut() preserves these structural fixed income intervals, whereas pd.qcut() would clump billionaires together with middle-class earners to keep sample counts equal.

Section C – Applied on Indian Startups Dataset

```python
import pandas as pd
import numpy as np

# Load dataset
df_raw = pd.read_csv('indian_startups_fundingDADS.csv')
df = df_raw.copy()

# 1. Rename columns to clean snake_case
df.columns = ['startup_name', 'city', 'industry', 'funding_stage', 'amount', 'funding_date', 'investors', 'founded_year']

# 2. Clean text column cases
df['startup_name'] = df['startup_name'].astype(str).str.strip().str.title()
df['city'] = df['city'].astype(str).str.strip().str.title()

# 3. Standardize city aliases
city_mapping = {
    'Bangalore': 'Bengaluru', 'Blr': 'Bengaluru',
    'Bombay': 'Mumbai', 'Bom': 'Mumbai',
    'Gurgaon': 'Gurugram', 'Ggn': 'Gurugram',
    'Delhi': 'New Delhi', 'Calcutta': 'Kolkata',
    'Madras': 'Chennai', 'Hyd': 'Hyderabad'
}
df['city'] = df['city'].replace(city_mapping)

# 4. Standardize industry and stage variants
ind_map = {
    'ecommerce': 'Ecommerce', 'e-commerce': 'Ecommerce',
    'beautytech': 'Beauty Tech', 'beauty-tech': 'Beauty Tech',
    'agritech': 'Agri Tech', 'agri-tech': 'Agri Tech', 'agri tech': 'Agri Tech',
    'edtech': 'Edtech', 'ed-tech': 'Edtech',
    'food tech': 'Food Tech', 'food-tech': 'Food Tech', 'foodtech': 'Food Tech',
    'saas': 'SAAS',
    'travel-tech': 'Travel Tech', 'traveltech': 'Travel Tech',
    'ai ml': 'AI ML', 'aiml': 'AI ML', 'ai/ml': 'AI ML', 'ai-ml': 'AI ML',
    'health-tech': 'Health Tech', 'healthtech': 'Health Tech', 'health tech': 'Health Tech',
    'electric-vehicles': 'Electric Vehicles', 'electric vehicles': 'Electric Vehicles', 'ev': 'Electric Vehicles',
    'fin-tech': 'Fintech', 'fintech': 'Fintech', 'ft': 'Fintech',
    'logistics': 'Logistics', 'proptech': 'Proptech', 'prop-tech': 'Proptech',
    'gaming': 'Gaming', 'games': 'Gaming'
}
df['industry'] = df['industry'].str.strip().str.lower().map(ind_map)

stage_map = {
    'series-e': 'Series E', 'series e': 'Series E', 'series_e': 'Series E',
    'pre-a': 'Pre-A', 'pre-series a': 'Pre-Series A', 'pre series a': 'Pre-Series A',
    'series-d': 'Series D', 'series d': 'Series D', 'series-c': 'Series C', 'series c': 'Series C',
    'series-b': 'Series B', 'series b': 'Series B', 'series_b': 'Series B',
    'series-a': 'Series A', 'series a': 'Series A', 'series_a': 'Series A',
    'ipo': 'IPO', 'pre-ipo': 'Pre-IPO', 'pre-seed': 'Pre-Seed', 'pre seed': 'Pre-Seed',
    'preseed': 'Pre-Seed', 'seed': 'Seed', 'bridge': 'Bridge'
}
df['funding_stage'] = df['funding_stage'].str.strip().str.lower().map(stage_map)

# 5. Clean money column
def clean_amount_value(val):
    if pd.isna(val): return np.nan
    s = str(val).strip().replace(',', '')
    if s.lower() in ['undisclosed', 'na', '']: return np.nan
    s = s.replace('INR', '').replace('₹', '').strip()
    s = s.lower().replace('crore', '').replace('cr', '').strip()
    try: return float(s)
    except: return np.nan

df['amount'] = df['amount'].apply(clean_amount_value)

# 6. Convert date column
df['funding_date'] = pd.to_datetime(df['funding_date'], errors='coerce', dayfirst=True)

# 7. Deduplicate records
df = df.drop_duplicates()
```
